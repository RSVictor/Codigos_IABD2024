```python
import pandas as pd #importa a biblioteca pandas
import matplotlib.pyplot as plt # importa a biblioteca matplotlib 
import numpy as np # importa a biblioteca numpy

```

```python
#Gerando numeros com a numpy
dados_x = np.linspace(0,10,10) #gera valores com a bilblioteca numpy
dados_y = dados_x * dados_x #gera valores multiplicando um dado pelo outro
```

```python
plt.figure(figsize=(10,5),facecolor='gray')
Ax = plt.axes() #cria os eixos dos grafico
Ax.set_facecolor('white') #cria o fundo do grafico
plt.xlabel('Eixo X',
           fontweight='bold',#nome do label
           fontsize='large',#estilo
           fontfamily='fantasy',#tamanho da fonte
           color='black',#cor da fonte
           )
plt.ylabel('Eixo Y',
           fontweight='bold',#nome do label
           fontsize='large',#estilo
           fontfamily='fantasy',#tamanho da fonte
           color='black',#cor da fonte
           )
plt.title("Exemplo Matplotlib",
          fontsize=12)#tamnho da fonte
plt.plot(dados_x,dados_y,
         color ='red',linewidth=2,
         marker= 'o',linestyle = 'dashed',markeredgecolor ='black')
```

```python
#Grafico 1
plt.rcParams.update({'font.size':15})#define o tamnho da fonte
plt.figure(figsize=(12,12))#define o tamnho da figura
linhas = 2#define o numero de linhas
colunas =2 #define o numero de colunas
plt.subplot(linhas,colunas,1)#define o numero de linhas e colunas
plt.plot(dados_x,'r--')
plt.title("Grfico 1")
#Grafico 2
plt.subplot(linhas,colunas,2)
plt.plot(dados_x*dados_y,'b-.')
plt.title('Grafico 2')
#Grafico 3
plt.subplot(linhas,colunas,3)
plt.plot(dados_x/2,'c-.')
plt.title("Grafico 3")
#Grafico 4 
plt.subplot(linhas,colunas,4)
plt.plot(dados_x, 'c-.')
plt.title("Grafico 4")
```
