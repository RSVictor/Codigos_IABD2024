```python
# Exercicio 3 
import numpy as np # importa a biblioteca numpy
from sklearn.linear_model import LinearRegression # da biblioteca sklearn importa o modelo de regressao linear
import matplotlib.pyplot as plt
```

```python
A = np.array([50,100,120,300]).reshape(-1,1)
Valor = np.array([180.000,300.000,375.000,600.000]).reshape(-1,1)
modelo = LinearRegression()
modelo.fit(A,Valor)
```

```python
A_teste = np.array([100,150,180,190]).reshape(-1,1)
valor_previsto= modelo.predict(Valor)
```

```python
for i in range(len(A_teste)): #Laço de repetição para exibir os dados de saida
    print(f"Área: {A_teste[i][0]} Valor previsto previsto:{valor_previsto[i]}")
    #gráfico de pontos -scatter
plt.scatter(A,Valor, color='blue' , label='Dados de Treinamento')
    #gráfico de linhas
plt.plot(Valor,valor_previsto, color = 'red', label='Teste')
plt.xlabel('Área')
plt.ylabel('Valor')
plt.title('Previsão de valor a partir da Área.')
plt.legend()
plt.grid(True)
plt.show
```
