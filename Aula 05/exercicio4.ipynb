```python
# Exercicio 4 
import numpy as np # importa a biblioteca numpy
from sklearn.linear_model import LinearRegression # da biblioteca sklearn importa o modelo de regressao linear
import matplotlib.pyplot as plt
```

```python
renda = np.array([1200,5000,8000,10000,30000]).reshape(-1,1)
gastos = np.array ([750,1500,4000,6000,18000]).reshape(-1,1)
modelo = LinearRegression()
modelo.fit(renda,gastos)
```

```python
renda_teste = np.array([1500,6500,9500,15000,45000]).reshape(-1,1)
gastos_previstos = modelo.predict(gastos)
```

```python
for i in range(len(renda_teste)): #Laço de repetição para exibir os dados de saida
    print(f"Renda: {renda_teste[i][0]} Gastos previsto:{gastos_previstos[i]}")
    #gráfico de pontos -scatter
plt.scatter(renda,gastos, color='blue' , label='Dados de Treinamento')
    #gráfico de linhas
plt.plot(gastos,gastos_previstos, color = 'red', label='Teste')
plt.xlabel('Renda')
plt.ylabel('Gastos')
plt.title('Previsão de gastos apartir da renda.')
plt.legend()
plt.grid(True)
plt.show
```
