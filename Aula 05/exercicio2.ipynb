```python
# Exercicio 2 
import numpy as np # importa a biblioteca numpy
from sklearn.linear_model import LinearRegression # da biblioteca sklearn importa o modelo de regressao linear
import matplotlib.pyplot as plt
```

```python
qnt=  np.array([100,200,300,400,500,600,700]).reshape(-1,1)
temp = np.array([30,35,40,45,60,100,150]).reshape(-1,1)
modelo = LinearRegression()
modelo.fit(qnt,temp)
```

```python
qnt_teste = np.array([150,250,350,450,550,650,800,]).reshape(-1, 1)
temp_prevista= modelo.predict(temp)
```

```python
for i in range(len(qnt_teste)): 
    print(f"Quantidade: {qnt_teste[i][0]} Temperatura prevista:{temp_prevista[i]} °C")
    #gráfico de pontos -scatter
plt.scatter(qnt,temp, color='blue' , label='Dados de Treinamento')
    #gráfico de linhas
plt.plot(temp,temp_prevista, color = 'red', label='Teste')
plt.xlabel('Quantidade')
plt.ylabel('Temperatura em °C')
plt.title('Previsão de temperatura a partir da Quantidade de calor.')
plt.legend()
plt.grid(True)
plt.show
```
