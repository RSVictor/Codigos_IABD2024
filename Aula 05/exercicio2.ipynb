```python
# Exercicio 2 
import numpy as np # importa a biblioteca numpy
from sklearn.linear_model import LinearRegression # da biblioteca sklearn importa o modelo de regressao linear
import matplotlib.pyplot as plt
```

```python
qnt=  np.array([100,200,300,400,500,600,700]).reshape(-1,1)
temp = np.array([30,35,40,45,60,100,150]).reshape(-1,1)
modelo = LinearRegression()
modelo.fit(qnt,temp)
```

```python
Q_teste = np.array([90,190,307,440,510,602,690]).reshape(-1, 1)
T_prevista= modelo.predict(Q_teste)
```

```python
for i in range(len(Q_teste)): 
    print(f"Calor: {Q_teste[i][0]} Temperatura prevista:{T_prevista[i]} °C")
    #gráfico de pontos -scatter
plt.scatter(qnt,temp, color='blue' , label='Dados de Treinamento')
    #gráfico de linhas
plt.plot(Q_teste,T_prevista, color = 'red', label='Teste')
plt.xlabel('Quantidade Calor')
plt.ylabel('Temperatura ')
plt.title('Previsão de temperatura a partir da Quantidade de calor.')
plt.legend()
plt.grid(True)
plt.show
```
