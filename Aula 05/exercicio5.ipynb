```python
# Exercicio 5 
import numpy as np # importa a biblioteca numpy
from sklearn.linear_model import LinearRegression # da biblioteca sklearn importa o modelo de regressao linear
import matplotlib.pyplot as plt
```

```python
consumo = np.array([30,50,75,123,140]).reshape(-1, 1)
demanda = np.array([80,100,100,160,200]).reshape(-1, 1)
modelo = LinearRegression()
modelo.fit(consumo,demanda)
```

```python
consumo_teste = np.array([100,80,200,30,500]).reshape(-1, 1)
demanda_prevista = modelo.predict(demanda)
```

```python
for i in range(len(consumo_teste)):
    print(f"Consumo: {consumo_teste[i][0]} Demanda prevista:{demanda_prevista[i]}")
    #gráfico de pontos -scatter
plt.scatter(consumo,demanda, color='blue' , label='Dados de Treinamento')
    #gráfico de linhas
plt.plot(consumo,demanda_prevista, color = 'red', label='Teste')
plt.xlabel('Consumo')
plt.ylabel('Demanda')
plt.title('Previsão de demanda com base no consumo.')
plt.legend()
plt.grid(True)
plt.show
```
